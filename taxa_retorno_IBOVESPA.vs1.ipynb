### Taxa de Retorno do IBOVESPA (^BVSP)

#### Bibliotecas

```python
#!pip install pandas_datareader
import numpy as np
import pandas as pd
from pandas_datareader import data as pdr
import matplotlib.pyplot as plt
```

```python
import yfinance as yf
yf.pdr_override()
ticker = ['^BVSP']
# definir data de início (start)
ibov_data = pdr.get_data_yahoo(ticker, start='2013-1-1', progress=False) ['Adj Close']
#ibov_data (apenas ativar se quiser vizualizar os dados carregados)
```

```python
# ativar apenas se desejar explorar os dados
#ibov_data.head()
#ibov_data.tail()
```

#### Normalização com base 100:

$$
\frac {P_t}{P_0} * 100
$$

```python
ibov_data.iloc[0]
```

```python
(ibov_data / ibov_data.iloc[0] * 100).plot(figsize = (15, 6));
plt.show()
```

#### Calculando a Taxa de Retorno do IBOVESPA

```python
returns = (ibov_data / ibov_data.shift(1)) - 1
returns.tail()
```

```python
annual_returns = returns.mean() * 250
annual_returns
```

```python
taxa_retorno_anual_IBOVESPA = str(round(annual_returns, 5) * 100) + '%'
print(taxa_retorno_anual_IBOVESPA)
```
