```python
#!pip install pandas_datareader
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from pandas_datareader import data as pdr
yf.pdr_override()
ticker = input('Digite o código da ação: ')
# definir data de início (start)
dados = pdr.get_data_yahoo(ticker, start='2022-1-1', progress=False)
```

## Taxa de Retorno Simples

$$
\frac{P_1 - P_0}{P_0} = \frac{P_1}{P_0} - 1
$$

```python
dados['simple_return'] = (dados['Adj Close'] / dados['Adj Close'].shift(1)) - 1
print (dados['simple_return'])
```

```python
dados['simple_return'].plot(figsize=(8, 5))
plt.show()
```

```python
avg_returns_d = dados['simple_return'].mean()
avg_returns_d
```

```python
avg_returns_a = dados['simple_return'].mean()*250
avg_returns_a
```

```python
print (str(round(avg_returns_a, 5)*100)+'%')
```

## Taxa de Retorno Logarítimo

$$
ln(\frac{P_t}{P_{t-1}})
$$

```python
dados['log_return'] = np.log(dados['Adj Close'] / dados['Adj Close'].shift(1))
print (dados['log_return'])
```

```python
dados['log_return'].plot(figsize=(8, 5))
plt.show()
```

```python
log_return_d = dados['log_return'].mean()
log_return_d
```

```python
log_return_a = dados['log_return'].mean()*250
log_return_a
```

```python
print (str(round(log_return_a, 5)*100)+'%')
```
